Scatter Plot: Used to display the relationship between two numeric variables. It helps identify patterns, correlations, and outliers.

Line Plot: Shows the trend of a continuous variable over a continuous interval or time. It is useful for visualizing data trends and patterns
.
Bar Plot: Used to compare categories or groups. It displays categorical data with rectangular bars, where the height represents the value of each category.

Histogram: Represents the distribution of a single numerical variable. It divides the data into bins and displays the frequency or count of observations in each bin.

Box Plot: Summarizes the distribution of a numeric variable through quartiles, outliers, and potential skewness. It helps identify central tendency, variability, and potential outliers.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the fmri dataset
fmri = sns.load_dataset("fmri")

# Plot line plot using x = "timepoint" and y = "signal" for different events and regions
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region")

# Display the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the titanic dataset
titanic = sns.load_dataset("titanic")

# Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'
plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
sns.boxplot(data=titanic, x="pclass", y="age")
plt.subplot(2, 1, 2)
sns.boxplot(data=titanic, x="pclass", y="fare")

# Display the plots
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the diamonds dataset
diamonds = sns.load_dataset("diamonds")

# Plot histogram for 'price' column with hue for 'cut' column
sns.histplot(data=diamonds, x="price", hue="cut")

# Display the plot
plt.show()
```

```python
import seaborn as sns

# Load the iris dataset
iris = sns.load_dataset("iris")

# Plot pair plot with hue for "species" column
sns.pairplot(data=iris, hue="species")

# Display the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the flights dataset
flights = sns.load_dataset("flights")

# Pivot the data to create a matrix form for the heatmap
flights_matrix = flights.pivot("month", "year", "passengers")

# Plot a heatmap
sns.heatmap(flights_matrix, cmap="YlGnBu")

# Display the plot
plt.show()
```
